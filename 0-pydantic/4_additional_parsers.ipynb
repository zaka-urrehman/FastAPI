```python
from pydantic import BaseModel, EmailStr

class Employee(BaseModel):
    name:str
    email: EmailStr

employee1 = Employee(name="abc" , email="abc@example.com")
```

```python
from pydantic import HttpUrl

class Employee2(BaseModel):
    name:str 
    email: EmailStr
    website: HttpUrl

employee2 = Employee2(name="abc" , email="abc@example.com" , website="https://example.com")
```

```python
from pydantic import PositiveInt

class Employee3(BaseModel):
    name:str
    age:PositiveInt


employee3=Employee3(name="abc" , age=20)
```
